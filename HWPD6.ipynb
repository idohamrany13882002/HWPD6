```python
import numpy as py
from IPython.display import display
import pandas as pd
```

```python
df = pd.read_csv("sport_athletes.csv")
```

```python
df
```

### Question 1

```python
results = df.copy()
results = results.groupby("Sport Type").agg({'Weight(kg)' : 'min', 'Height(cm)' : 'max', 'Wins' : 'max'})
results.rename(columns = {'Weight(kg)' : 'min_Weight(kg)', 'Height(cm)' : 'Max_Height(cm)', 'Wins' : 'Max_Wins'}, inplace = True)
display(results)
```

### Question 2

```python
tallest2 = df.groupby('Sport Type')['Height(cm)'].nlargest(2).reset_index()
display(tallest2)
```

### Question 3

```python
grouped = df.groupby(["Country", "Sport Type"]).max()
display(grouped)
```

```python
basketball_usa = grouped.loc[[("USA", "Basketball")]]
display(basketball_usa)
```

```python
soccer_stats = grouped.xs("Soccer", level = "Sport Type")
display(soccer_stats)
```

```python
soccer_tennis = grouped[grouped.index.get_level_values("Sport Type").isin(["Soccer", "Tennis"])]
display(soccer_tennis)
```

```python
swapped = grouped.swaplevel().sort_index()
display(swapped)
```

```python
stats = df[["Wins", "Height(cm)"]].agg(['min', 'max', 'mean', 'std'])
display(stats)
```
